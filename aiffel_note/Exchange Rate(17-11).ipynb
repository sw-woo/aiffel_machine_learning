```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#URL 가져오기 200은 성공이라는 뜻
url = 'https://m.stock.naver.com/marketindex/index.nhn'
response = requests.get(url)
response
```

```python
#Soup 객체 생성
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#원하는 데이터 추출 - 국가
country = []
country_1 = soup.findAll('strong')
country_2 = soup.findAll('span', {'class':'stock_item'})
for c in country_1:
    country.append(c.text)
for c in country_2:
    country.append(c.text)
country.remove('증권')
country
```

```python
#원하는 데이터 추출 - 환율정보
price = []
price_0 = soup.findAll('span', {"class":"stock_price"})
for p in price_0:
    price.append(p.text)

price
```

```python
#데이터 정렬, (pandas)
import pandas as pd
data = {}
sum = 0
for i in range(len(price)):
    data[country[i]]=price[i]
pd.Series(data)
```

```python
index_exchange = pd.DataFrame(country)
index_exchange
```

```python
index_exchange2 = pd.DataFrame(price)
index_exchange2
```

```python
result = pd.concat([index_exchange,index_exchange2], axis=1)
result
```
